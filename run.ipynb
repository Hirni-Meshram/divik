```python
from multiprocessing import Pool

import numpy as np
import pandas as pd
from sklearn.externals import joblib
from tqdm import tqdm_notebook

import spdivik.distance as dst
import spdivik.predefined


if __name__ == '__main__':
    metadata = pd.read_csv(meta)
    data = np.load(source)
    data = data[metadata.dataset == 1]
    print(data.shape)
    distance = dst.ScipyDistance(dst.KnownMetric.correlation)
    with Pool(maxtasksperchild=4) as pool, tqdm_notebook(desc='processed', total=data.shape[0]) as progress:
        result = spdivik.predefined.master(pool=pool, progress_reporter=progress, distance=distance)(data)
    joblib.dump(result, destination)
```
